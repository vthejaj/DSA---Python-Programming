```python
def func_new(a):
  print("Value of a : -->",a)
  if a<=0:
    return a
  else:
    return func_new(a-1)

ret_val = func_new(5)
print("return val: -->", ret_val)
```

```python
def factorial(n):
  print("value of n: -->",n)
  if n==0:
    return 1
  return n*factorial(n-1)

ret_val = factorial(5)
print("ret val: -->", ret_val)
```

```python
def sum_n_numbers(n):
  print("n val : -->",n)
  if n==0:
    return 0
  return n+sum_n_numbers(n-1)

ret_val = sum_n_numbers(5)
print(ret_val)
```

```python
def head_recursion(n):
  if n>0:
    head_recursion(n-1) # head_recursion(1-1)
    print("n val: -->", n) # print(1)

# head_recursion(5)--->head_recursion(1)

def tail_recursion(n):
  if n>0:
    print("n val: -->", n)
    tail_recursion(n-1)
```

```python
head_recursion(5)
```

```python
tail_recursion(5)
```

```python
def normal_func(n):
  if n>0:
    print(1+5)

normal_func(0)
```

```python
def finite_recursion(n):
  print("val of n: -->", n)
  if n==0:
    return 0
  return finite_recursion(n-2) # direct recursion
```

```python
finite_recursion(6)
```

```python
finite_recursion(7)
```

```python
# indirect recursion

def func_A(n):
  if n>0:
    print("In func A: -->", n)
    func_B(n//5)

def func_B(n):
  if n>0:
    print("In func B: -->", n)
    func_A(n-1)
```

```python
func_A(20)
```

```python
# nested recursions

def nested_recursion(n):
  if n>100:
    return n-10
  return nested_recursion(nested_recursion(n+11))

nested_recursion(95)
```

```python
# you can have more than 1 function calls in a recursive function

# 1, 1, 2, 3, 5 , 8, 13, 21
# 1  2  3  4   5  6  7   8

# f(n) = f(n-1)+f(n-2)

def getFibonacciNum(n):
  if n==1 or n==2:
    return 1
  return getFibonacciNum(n-1)+getFibonacciNum(n-2) # getFibonacciNum(7) + getFibonacciNum(6)
```

```python
getFibonacciNum(8)
```
